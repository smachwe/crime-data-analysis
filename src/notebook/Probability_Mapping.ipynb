```python
import numpy as np
import pandas as pd
from ast import literal_eval
```

```python
ref = pd.read_csv('./ref_class_probs.csv')
profiles = pd.read_csv('./user_profilesv3.csv', converters={'att': literal_eval}, usecols = range(1,27))
```

```python
def prob_generator(crime_type):
    att_dict = {}
    for index, row in ref.iterrows():
        att_dict[row['class']] = row[crime_type]

    profile_dict = {}
    for index, value in enumerate(profiles['att']):
        profile_dict[index] = value

    for item,value in profile_dict.items():
        probs = []
        for v in value:
            probs.append(att_dict[v])
        profile_dict[item] = probs
    probs_dict = {k:np.prod(v) for k,v in profile_dict.items()}
    return pd.Series(probs_dict)
```

```python
crime_types = ['rape_sexual_assault','robbery','battery','assault'
               ,'burglary_tresspassing','motor_vehicle_theft','theft']
for crime_type in crime_types:
    profiles[crime_type] = (prob_generator(crime_type))
```

```python
profiles.to_csv('./profile_probs.csv')
```
