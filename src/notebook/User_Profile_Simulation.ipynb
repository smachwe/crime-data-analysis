1. Description - The primary use for this notebook is to create a simulation of user profiles for the Chicago Crime Dataset

2. Import packages & Load data

```python
import pandas as pd
import numpy as np
```

```python
## Df of potential features for user profiles
df = pd.read_excel('Victim_Profile_Categories _v2.xlsx', skiprows=1, header = 1)
df
```

```python
df.iloc[0,1:]
```

3. Create Feature Dictionary - The below column maps out which features map to what columns right side excluded

```python
feature_dict = {'Gender' : (0,2),
                'Age' : (2,8),
                'Hispanic': (8,10),
                'Income': (10,17),
                'Marital_Status': (17,22),
                'Race': (22,25),
                }
```

```python
random_tuples = list(feature_dict.values())
random_tuples
```

4. Simulation - In the below code the steps for simulation are as follows: 

      *   creating a dataframe called user_profiles
      *   creating an zero filled array of length equal to user_profile     columns   
      *   Choosing one column from each of the features in feature dict and setting that column value to 1

      *   appending new array to user_profiles



```python
user_profiles = pd.DataFrame(columns=df.columns[1:])
iterations = 1000
for i in range(iterations):
  zero_array = np.zeros(len(df.iloc[0,1:]))
  for col in random_tuples:
    zero_array[np.random.randint(col[0],col[1])] = 1
  user_profiles.loc[len(user_profiles)] = zero_array  
  print(i+1,'User_profiles Created')
```

```python
user_profiles
```

5. Graphs & results - In the following cells we will save the output of user_profiles to a csv file as well as display a histogram of each variable.

```python
attributes = []
for i in range(len(user_profiles)):
  transposed_col = user_profiles.iloc[i:i+1,:].T
  transposed_col.columns = ['att']
  attributes.append(list(transposed_col[transposed_col.att == 1].index))
att = pd.DataFrame({'att':attributes})
```

```python
user_profiles = pd.concat([user_profiles,att],axis = 1)
```

```python
user_profiles.to_csv('user_profilesv3.csv')
```

```python
import matplotlib.pyplot as plt
for col in user_profiles.columns:
  print('---------------',col,'---------------')
  plt.hist(user_profiles.loc[:,col])
  plt.pause(.05)
```
